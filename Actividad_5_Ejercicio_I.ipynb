```python
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.optimizers import Adam

# Ejercicio 1: Carga del Conjunto de Datos
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

# Ejercicio 2: Normalización de los Datos
x_train = x_train / 255.0
x_test = x_test / 255.0

# Ejercicio 3: Construcción del Modelo
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

# Ejercicio 4: Compilación del Modelo
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Ejercicio 5: Entrenamiento del Modelo
model.fit(x_train.reshape(-1, 28, 28, 1), y_train, epochs=10)

# Ejercicio 6: Evaluación del Modelo
train_loss, train_accuracy = model.evaluate(x_train.reshape(-1, 28, 28, 1), y_train)
test_loss, test_accuracy = model.evaluate(x_test.reshape(-1, 28, 28, 1), y_test)

print("Pérdida y precisión en el conjunto de entrenamiento:")
print("Pérdida:", train_loss)
print("Precisión:", train_accuracy)
print("\nPérdida y precisión en el conjunto de prueba:")
print("Pérdida:", test_loss)
print("Precisión:", test_accuracy)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import cifar10, cifar100
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from keras.optimizers import Adam
from keras.utils import to_categorical

# Ejercicio 1: Carga de los conjuntos de datos
(x_train_cifar10, y_train_cifar10), (x_test_cifar10, y_test_cifar10) = cifar10.load_data()
(x_train_cifar100, y_train_cifar100), (x_test_cifar100, y_test_cifar100) = cifar100.load_data()

# Ejercicio 2: Normalización de los datos
x_train_cifar10 = x_train_cifar10.astype('float32') / 255
x_test_cifar10 = x_test_cifar10.astype('float32') / 255
x_train_cifar100 = x_train_cifar100.astype('float32') / 255
x_test_cifar100 = x_test_cifar100.astype('float32') / 255

# Ejercicio 3: Construcción del modelo
model_cifar10 = Sequential()
model_cifar10.add(Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model_cifar10.add(MaxPooling2D((2, 2)))
model_cifar10.add(Conv2D(64, (3, 3), activation='relu'))
model_cifar10.add(MaxPooling2D((2, 2)))
model_cifar10.add(Flatten())
model_cifar10.add(Dense(128, activation='relu'))
model_cifar10.add(Dense(10, activation='softmax'))

model_cifar100 = Sequential()
model_cifar100.add(Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model_cifar100.add(MaxPooling2D((2, 2)))
model_cifar100.add(Conv2D(64, (3, 3), activation='relu'))
model_cifar100.add(MaxPooling2D((2, 2)))
model_cifar100.add(Flatten())
model_cifar100.add(Dense(128, activation='relu'))
model_cifar100.add(Dense(100, activation='softmax'))

# Ejercicio 4: Compilación del modelo
model_cifar10.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model_cifar100.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Ejercicio 5: Entrenamiento del modelo
history_cifar10 = model_cifar10.fit(x_train_cifar10, y_train_cifar10, epochs=10, batch_size=128, validation_data=(x_test_cifar10, y_test_cifar10))
history_cifar100 = model_cifar100.fit(x_train_cifar100, y_train_cifar100, epochs=10, batch_size=128, validation_data=(x_test_cifar100, y_test_cifar100))

# Ejercicio 6: Evaluación del modelo
train_loss_cifar10, train_accuracy_cifar10 = model_cifar10.evaluate(x_train_cifar10, y_train_cifar10, verbose=0)
test_loss_cifar10, test_accuracy_cifar10 = model_cifar10.evaluate(x_test_cifar10, y_test_cifar10, verbose=0)
train_loss_cifar100, train_accuracy_cifar100 = model_cifar100.evaluate(x_train_cifar100, y_train_cifar100, verbose=0)
test_loss_cifar100, test_accuracy_cifar100 = model_cifar100.evaluate(x_test_cifar100, y_test_cifar100, verbose=0)

print("CIFAR-10:")
print("Pérdida y precisión en el conjunto de entrenamiento:", train_loss_cifar10, train_accuracy_cifar10)
print("Pérdida y precisión en el conjunto de prueba:", test_loss_cifar10, test_accuracy_cifar10)

print("\nCIFAR-100:")
print("Pérdida y precisión en el conjunto de entrenamiento:", train_loss_cifar100, train_accuracy_cifar100)
print("Pérdida y precisión en el conjunto de prueba:", test_loss_cifar100, test_accuracy_cifar100)

# Ejercicio 7: Visualización de resultados
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history_cifar10.history['accuracy'], label='CIFAR-10 Training Accuracy')
plt.plot(history_cifar10.history['val_accuracy'], label='CIFAR-10 Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history_cifar10.history['loss'], label='CIFAR-10 Training Loss')
plt.plot(history_cifar10.history['val_loss'], label='CIFAR-10 Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.show()

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history_cifar100.history['accuracy'], label='CIFAR-100 Training Accuracy')
plt.plot(history_cifar100.history['val_accuracy'], label='CIFAR-100 Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history_cifar100.history['loss'], label='CIFAR-100 Training Loss')
plt.plot(history_cifar100.history['val_loss'], label='CIFAR-100 Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.show()

# Ejercicio 7: Visualización de ejemplos de imágenes de prueba
def plot_example_images(images, labels, predictions=None):
    plt.figure(figsize=(10, 10))
    for i in range(25):
        plt.subplot(5, 5, i+1)
        plt.xticks([])
        plt.yticks([])
        plt.grid(False)
        plt.imshow(images[i])
        if predictions is not None:
            plt.xlabel("Real: {} / Predicción: {}".format(labels[i], predictions[i]))
        else:
            plt.xlabel("Real: {}".format(labels[i]))
    plt.show()

predictions_cifar10 = model_cifar10.predict_classes(x_test_cifar10)
predictions_cifar100 = model_cifar100.predict_classes(x_test_cifar100)

plot_example_images(x_test_cifar10, y_test_cifar10, predictions_cifar10)
plot_example_images(x_test_cifar100, y_test_cifar100, predictions_cifar100)
```
